```python
import pandas as pd
import matplotlib.pyplot as plt

# Charge le CSV
df = pd.read_csv('dev/climate_data/cdu_cac40_cleaned.csv')

# Choisir l'entrepris
entreprise = "TotalEnergies SE"
row = df[df['Company'].str.contains("Total", case=False)].iloc[0]

# Années disponibles
annees = [2018, 2019, 2020, 2021, 2022]
valeurs = [row[str(a)] if str(a) in row.index else None for a in annees]

# Nettoyer les valeurs (enlever les virgules)
valeurs = [float(str(v).replace(',', '')) if pd.notna(v) and str(v).strip() != '' else None for v in valeurs]

# Graphique 1 — Évolution temporelle
plt.figure(figsize=(10, 5))
plt.plot(annees, valeurs, marker='o', color='#378ADD', linewidth=2)
plt.fill_between(annees, valeurs, alpha=0.15, color='#378ADD')
plt.title(f'Évolution des émissions — {row["Company"]}')
plt.ylabel('tCO₂e')
plt.xlabel('Année')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('emissions_evolution.png', dpi=150)
plt.show()

# Graphique 2 — Scope 1 vs Scope 2
scope1 = float(str(row['Scope 1']).replace(',', ''))
scope2 = float(str(row['Scope 2 (location-based)']).replace(',', ''))

plt.figure(figsize=(6, 5))
plt.bar(['Scope 1', 'Scope 2'], [scope1, scope2], color=['#378ADD', '#9FE1CB'])
plt.title(f'Répartition des émissions — {row["Company"]}')
plt.ylabel('tCO₂e')
plt.tight_layout()
plt.savefig('emissions_scope.png', dpi=150)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dev/climate_data/cdu_cac40_cleaned.csv', skiprows=2)

entreprise = "TotalEnergies SE"
row = df[df['Company'].str.contains("Total", case=False)].iloc[0]

annees = [2018, 2019, 2020, 2021, 2022]
valeurs = []
for a in annees:
    v = row[str(a)]
    if pd.notna(v) and str(v).strip() != '':
        valeurs.append(float(str(v).replace(',', '')))
    else:
        valeurs.append(None)

# Graphique 1 — Évolution temporelle
plt.figure(figsize=(10, 5))
plt.plot(annees, valeurs, marker='o', color='#378ADD', linewidth=2)
plt.fill_between(annees, valeurs, alpha=0.15, color='#378ADD')
plt.title(f'Évolution des émissions — {row["Company"]}')
plt.ylabel('tCO₂e')
plt.xlabel('Année')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('emissions_evolution.png', dpi=150)
plt.show()

# Graphique 2 — Scope 1 vs Scope 2
scope1 = float(str(row['Scope 1']).replace(',', ''))
scope2 = float(str(row['Scope 2 (location-based)']).replace(',', ''))

plt.figure(figsize=(6, 5))
plt.bar(['Scope 1', 'Scope 2'], [scope1, scope2], color=['#378ADD', '#9FE1CB'])
plt.title(f'Répartition des émissions — {row["Company"]}')
plt.ylabel('tCO₂e')
plt.tight_layout()
plt.savefig('emissions_scope.png', dpi=150)
plt.show()
```
